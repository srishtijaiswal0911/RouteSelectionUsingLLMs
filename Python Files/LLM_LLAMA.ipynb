```python
embeddings = HuggingFaceEmbeddings()
```

```python
index_name = "projectllm"
index = pinecone.Index(index_name)
vectordb = Pinecone.from_documents(texts, embeddings, index_name=index_name) #Storing the text in vectorDB format
```

```python
llm = Replicate( #Llama Model
    model="a16z-infra/llama13b-v2-chat:df7690f1994d94e96ad9d568eac121aecf50684a0b0963b25a41cc40061269e5",
    model_kwargs={"temperature": 0.75, "max_length": 3000}
) 
```

```python
qa_chain = ConversationalRetrievalChain.from_llm(
    llm,
    vectordb.as_retriever(search_kwargs={'k': 2}),
    return_source_documents=True
)
```

#### Sending prompts to the LLM

```python
chat_history = [] #For storing chat history
import time

start_time = time.perf_counter()
#Continuosly take user input
while True:
    query = input('Prompt: ')
    if query == "exit" or query == "quit" or query == "q":
        print('Exiting')
        sys.exit()
    result = qa_chain({'question': query, 'chat_history': chat_history})
    print('Answer: ' + result['answer'] + '\n')
    chat_history.append((query, result['answer']))
end_time = time.perf_counter()
response_time = end_time - start_time
print(f"Response Time: {response_time} seconds")
```

```python
import time

chat_history = []

while True:
    query = input('Prompt: ')
    if query.lower() in ["exit", "quit", "q"]:
        print('Exiting')
        break

    start_time = time.perf_counter()  # Start timer before LLM query
    result = qa_chain({'question': query, 'chat_history': chat_history})
    end_time = time.perf_counter()  # Stop timer after getting the response

    response_time = end_time - start_time  # Calculate response time for this query
    print(f'Answer: {result["answer"]}\n')
    print(f"Response Time for this query: {response_time} seconds\n")

    chat_history.append((query, result['answer']))
```
